```python
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy as scs
```

```python
def johleker(x,p):
    return(1-(((2*x)-1)**(2*p)))

def biolek(x,p,i):
    if -i >= 0:
        return(1-((x-1)**(2*p)))
    else:
        return(1-(x**(2*p)))

def prodromakis(x,p,j):
    dum=(x-0.5)**2
    dum1=(dum+0.75)**p
    return(j*(1-dum1))
```

```python
x=np.linspace(0,1,10000)
p=10
j=4

jo = []
bipos = []
bineg = []
pa = []

for i in range(len(x)):
    jo.append(johleker(x[i],p))
    bipos.append(biolek(x[i],p,1))
    bineg.append(biolek(x[i],p,-1))
    pa.append(prodromakis(x[i],p,j))
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.plot(x,jo,color='black',label='Jogleker')
plt.xlabel('x',fontsize=15)
plt.ylabel('f(x)',fontsize=15)
plt.grid()
plt.legend()
plt.subplot(2,2,2)
plt.plot(x,bipos,color='red',label='Biolek Posetive')
plt.xlabel('x',fontsize=15)
plt.ylabel('f(x)',fontsize=15)
plt.grid()
plt.legend()
plt.subplot(2,2,3)
plt.plot(x,bineg,color='blue',label='Biolek Negative')
plt.xlabel('x',fontsize=15)
plt.ylabel('f(x)',fontsize=15)
plt.grid()
plt.legend()
plt.subplot(2,2,4)
plt.plot(x,pa,color='green',label='Prodromakis')
plt.xlabel('x',fontsize=15)
plt.ylabel('f(x)',fontsize=15)
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=(20,10))
plt.plot(x,jo,color='black',label='Jogleker')
plt.plot(x,bipos,color='red',label='Biolek Posetive')
plt.plot(x,bineg,color='blue',label='Biolek Negative')
plt.plot(x,pa,color='green',label='Prodromakis',linestyle='--',linewidth=5)
plt.xlabel('x',fontsize=15)
plt.ylabel('f(x)',fontsize=15)
plt.grid()
plt.legend()
plt.show()
```

```python
def linear(roff,ron,woff,won,v,w):
    dum=(roff-ron)/(woff-won)
    dum2=w-won
    dum3=(ron+(dum*dum2))**(-1)
    return(dum3*v)
import math
    
def exponential(roff,ron,woff,won,v,w):
    dum = math.log(roff/ron)
    dum1=np.exp((-dum*(w-won))/(woff-won))
    return(dum1*v/ron)


def exponential(roff,ron,woff,won,v,w):
    dum = math.log(roff/ron)
    dum1 = (dum*(w-won))/(woff-won)
    dum2=np.exp(-dum1)
    return(dum2*v/ron)
```

```python
w = [1e-9,2e-9,3e-9,4e-9,5e-9]
roff = 100e3
ron = 100
woff =10e-9
won = 0
v = np.linspace(-10,1000,1000)

lin=[]
expo=[]
for j in range(len(w)):
    dum1=[]
    dum2=[]
    for i in range(len(v)):
        dum1.append(linear(roff,ron,woff,won,v[i],w[j]))
        dum2.append(exponential(roff,ron,woff,won,v[i],w[j]))
    lin.append(dum1)
    expo.append(dum2)

plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.title('Linear Relationship',fontsize=20)
for i in range(len(lin)):
    plt.plot(v,lin[i],color='black',label='w = '+str(i+1)+'nm')
    
plt.ylabel('Current (I)',fontsize=15)
plt.xlabel('Voltage (V)',fontsize=15)
plt.grid()
plt.legend()
plt.subplot(1,2,2)
plt.title('Exponential Relationship',fontsize=20)

for i in range(len(lin)):
    plt.plot(v,expo[i],color='black',label='w = '+str(i+1)+'nm')
plt.ylabel('Current (I)',fontsize=15)
plt.xlabel('Voltage (V)',fontsize=15)
plt.grid()
plt.legend()
plt.show()
```
