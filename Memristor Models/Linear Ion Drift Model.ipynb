```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ron = 100#10e3
Roff = 160e3#100e3
Xinit = 0.1#0.1
D = 10e-9#27e-9
muv = 1e-14
t = np.linspace(0,10,100000)
f = 0.1#0.5
amp = 1e-5
I = np.sin(2*np.pi*f*t)*amp
test_list = {'Ron':Ron,
            'Roff':Roff,
            'Xinit':Xinit,
            'D':D,
            'muv':muv}
```

```python
def memristor(muv, Ron, Roff, D, I, t, Xinit):
    
    def dx_func(muv, Ron, D, I, dt):
        dum = (muv*Ron)/(D**2)
        return(dum*I*dt)
    
    
    out = []
    
    x = Xinit
    
    for i in range(len(I)):
        if i == 0:
            out.append(0)
        else:
            dt = t[i]-t[i-1]
            v = ((Ron*x) + (Roff*(1-x)))*I[i]
            out.append(v)
            dx = dx_func(muv, Ron, D, I[i], dt)
            x = x+dx
    return(out)
```

```python
v = memristor(muv, Ron, Roff, D, I, t, Xinit)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.plot(v,I)
plt.ylabel('Current (A)',fontsize=15)
plt.xlabel('Voltage (V)',fontsize=15)
plt.grid()
plt.subplot(1,2,2)
plt.plot(v,np.abs(I))
plt.ylabel('Current (A)',fontsize=15)
plt.xlabel('Voltage (V)',fontsize=15)
plt.grid()
plt.show()

plt.plot(v,I)
plt.ylabel('Current (A)',fontsize=15)
plt.xlabel('Voltage (V)',fontsize=15)
plt.grid()
plt.show()
plt.plot(v,np.abs(I))
plt.ylabel('Current (A)',fontsize=15)
plt.xlabel('Voltage (V)',fontsize=15)
plt.grid()
plt.show()
```
